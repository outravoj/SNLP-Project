```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/SNLP projekt/data/train.csv").head(40_000)
```

```python
labels = df["is_duplicate"].to_list()
print(f"The ratio of positive labels is {sum(labels)/len(labels)}")
```

```python
print(df.columns)
```

```python
qid1, qid2 = max(df["qid1"].to_list()), max(df["qid2"].to_list())
print(f"Number of different questions is: {max(qid1,qid2)}")
```

```python
quest1 =  max([len(str(x).split()) for x in df["question1"].to_list()])
quest2 =  max([len(str(x).split()) for x in df["question2"].to_list()])
print(f"Length of the longest sentence is {max(quest1, quest2)}")
```

```python
quest1 =  df["question1"].to_list()
quest2 =  df["question2"].to_list()
pair_lens = [len((str(quest1[i]).split()+str(quest2[i]).split())) for i in range(len(quest1))]
sns.distplot(pair_lens)
plt.xlim([0, 256]);
plt.xlabel('Token count');
```
